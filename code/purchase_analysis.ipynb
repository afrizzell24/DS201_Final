# Climate Change Prediction

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

# Stage 2

### First, import and read the data

```python
df = pd.read_csv('../data/purchasing_behaviors.csv')

df.info()
df.head()
```

```python
df.describe()
```

### We have enough information to completely remove rows with missing data

```python
print(df.shape)
df = df.drop_duplicates()
df = df.dropna()
print(df.shape)
```

Our data was completely clean to start with!

### Data preparation

```python
df = df.drop(columns=['user_id', 'region'])
df.head()
```

# Stage 3

### Exploring the data (Analysis in slides)

```python
sns.pairplot(df, markers = '+', kind = 'reg', plot_kws = dict(line_kws = dict(color = 'red'), scatter_kws = dict(linewidths = 1)))
```

```python
corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], color='green', bins=10)

plt.title('Age Distribution', fontsize=16)
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
